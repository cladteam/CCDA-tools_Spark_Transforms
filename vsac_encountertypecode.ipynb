```python
from foundry.transforms import Dataset

ccda_encounter_code_value_set = Dataset.get("ccda_encounter_code_value_set")
ccda_encounter_code_value_set_files = ccda_encounter_code_value_set.files().download()
```

```python
print(ccda_encounter_code_value_set_files['vsac_EncounterTypeCode.json'])
```

```python
import json
import pandas as pd

# Load the JSON data
with open(ccda_encounter_code_value_set_files['vsac_EncounterTypeCode.json'], 'r') as file:
    data = json.load(file)

# Extract the relevant data from the 'expansion' -> 'contains' section
encounter_types = data['expansion']['contains']

# Extract the 'id' from the JSON root level
json_id = data['id']

# Create a DataFrame from the extracted data
df = pd.DataFrame(encounter_types)

# Add the 'id' column to the DataFrame
df['oid'] = json_id

# Select only relevant columns for clarity
df_vsac_encountertypecode = df[['oid', 'system', 'version', 'code', 'display']]

# Display the DataFrame or save it to a CSV file
print(df_vsac_encountertypecode)
```

```python
from foundry.transforms import Dataset

vsac_encountertypecode = Dataset.get("vsac_encountertypecode")
vsac_encountertypecode.write_table(df_vsac_encountertypecode)
```
