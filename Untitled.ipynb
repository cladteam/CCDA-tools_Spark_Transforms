```python
from foundry.transforms import Dataset

ccda_encounter_code_value_set = Dataset.get("ccda_encounter_code_value_set")
ccda_encounter_code_value_set_files = ccda_encounter_code_value_set.files().download()
```

```python
import json
import pandas as pd

# Load the JSON data
data = json.load(ccda_encounter_code_value_set)

# Extract the relevant data from the 'expansion' -> 'contains' section
encounter_types = data['expansion']['contains']

# Create a DataFrame from the extracted data
df = pd.DataFrame(encounter_types)

# Select only relevant columns for clarity
df = df[['system', 'version', 'code', 'display']]

# Display the DataFrame or save it to a CSV file
print(df)
#df.to_csv('EncounterTypeCode.csv', index=False)

print("Table has been saved to 'EncounterTypeCode.csv'")
```
